```python
import random

#Kullanıcı Yaşını Girecek
#1- 10 yaş arası 1-20 arası rastgele sayı üretilecek.
#10- 20 yaş arası 1-50 arası rastgele sayı üretilecek.
#20- 50 yaş arası 1-100 arası rastgele sayı üretilecek.

print("Haşmet Dayının pizza dükkanı hoşgeldiniz.")
print("Pizanın içinden çıkabilecekler böcek sayısını doğru tahmin edersiniz")
print("pizanın bedava!")
yas = input("Lütfen öence yaşınızı giriniz")
if (yas >= 1 and yas <=10):
  rastgele_sayi = random.randint(1,20)
  print("1-20 arasında sayı tahmine edeceksin")
if (yas >= 11 and yas <=20):
    rastgele_sayi = random.randint(1,50)
    print("1-100 arasında sayı tahmine edeceksin")
if (yas >= 21 and yas <=50):
  rastgele_sayi = random.randint(1,100)
  tahmin_hakkı = 5
  tahmin = int(input("lütfen tahminizi giriniz:"))
while (tahmin != rastgele_sayi):
  tahmin_hakkı -= 1
  print("Tahmininiz yanlış.",rastgele_sayı)
  tahmin = int(input("tahmininz yanlış.","tahmin hakınız kaldı. tekrar tahmin et"))
  if tahmin_hakkı == 0:
    print("Tahmin hakkınız bitti!")
    break #While döngüsü kırıyoruz.

  if (tahmin == rastgele_sayi):
      print("Tebrikler! Sayıyı Doqru Bildiniz!")

print("Böcek sayısının tahmin edemedinizz. Ücretiniz:900 Tl")








```
