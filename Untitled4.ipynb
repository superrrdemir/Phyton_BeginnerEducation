 # While Döngüsü
 While döngüsü, for'dan faklı olarak bir koşulun oluşması durumunda
 çalışır. En temelinde if ve for'un karışımı gibidir. Örneğin:

 1. Eğer a'nın değeri 40'tan küçük olduğu sürece çalışır,
 2. Eğer parola True olduğu sürece çalış.
 3. Eğer kulanıcı_adı "Haşmet Dayı" olduğu sürece çalış.

 While, her zaman bir koşula ihtiyaç duyar!

```python
while(True)
  print("Muhamed")
```

NOT: Yukarıdaki örnek sonsuza kadar çalışır. Fazla kullanmak sağlığa zararlardır. Bilgisiyar sağlığına zararlıdır

```python
sayi = 0
while(sayi<=10): #Sayi 10 a eşit ve 10'dan küçük olduğu sürece çalışır
  print(sayi)
  sayi+=1 #sayi 1'er 1'er sayı ekle
```

```python
bilgi = input("Kullanıcı Adınızı Yazınız:")
while(bilgi != "Haşmet Dayı"):
  print("Kulanıcı adı yanlış!")
  bilgi = input("Kullanıcı Adınızı Yazınız:")
  print("Giriş Başarılı! Hoşgeldin Yüce Haşmet Dayı!")
```

For Döngüsü İle yıldız ve üçgen

```python
for k in range(1,10,1):
  for i in range(k):
    print("*",end="")
  print()
for a in range(1,10,-1):
  for b in range(a):
    print("*",end="")
  print()
```
