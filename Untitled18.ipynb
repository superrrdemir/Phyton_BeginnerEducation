```python
# OS  Kütüphanesi verileri nasıl ayrıca çekeriz?
import os
hasmet_place = open("personel.txt","w")
personel_adı =[] # Haşmet Dayının yerindeki çalışan kişinin adı
personel_maas = [] # Personerleri aldığı ücretler

while True: #Ben kırana kadar çalışmayı devam et
  personel_bilgi = input("Lütfen çalışan adına giriniz: ")
  personel_para = int(input("Lütfen maaş girin(sayı): "))
  personel_adı.append(personel_bilgi)
  personel_maas.append(personel_para)
  if personel_bilgi == "0" or personel_para == 0:
   personel_adı.remove(personel_bilgi)
   personel_maas.remove(personel_para)
   break

 #Yazma işlemi bitince not defterine biligileri ekleyelim

for bilgi in personel_bilgi:
  for maas in personel_maas:
   verilerim = str(bilgi)+"_"+str(maas)+"TL"
  hasmet_place.writelines(str(verilerim))
hasmet_place.close()
```
