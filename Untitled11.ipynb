## Fonksyonna Parametre Yazma
Fonksyon çalışacak değişkenlere parametreler denir. Bu parametreler fonksyonun içinde çalışacak deperlerdir Ana programda yazılımcı zorunlu
olarak bunu doldurmalıdır İster doğrudan bir değer yazar isters elle gibi




```python
#Paremetler parantez içine yazılır.
def kulanici_adi(kullanici_adi, parola):
  if kullanici_adi == "Gary" and parola == "12a34b":
    print("Başarılı kullanıcı girişi!")
  else:
    print("Başarısız Giriş")

```

```python
k_adi = input("Kullanıcı adınızı giriniz:")
sifre = input("Parolanızı giriniz:")
kulanici_adi(k_adi, sifre)
```

```python
kulanici_adi("Gary", "12a34b")
```
