#Fonksyonlar - Metorlar

tabanla programlamlamanın temelidir. "Nesne Tabanlama Proglamlama (NTP)" İngilizce "Object Orintanied Programming (oop)" olarak bilinir. Bu,

işlemleri ve değişkeliri sakamyı amaçlar. <br>

2. Bir matamatik işlemi yapıyoruzdiyelim .Bu fonksilyon içine saklayıp o fonkisyonu çağırarak satıralrca kod yazılmaca engel olur
