```python
import math

while True:
    print("\nHesaplamak istediğiniz alan türünü seçin:")
    print("1: Dikdörtgen")
    print("2: Kare")
    print("3: Üçgen")
    print("4: Paralelkenar")
    print("5: Silindir")
    print("6: Daire")
    print("0: Çıkış")

    secim = input("Seçiminizi girin: ")

    if secim == '0':
        break
    elif secim == '1':
        print("Dikdörtgenin alan")
        Dikdörtgen_kısa = int(input("Dikdörtgenin kısa kenarını giriniz: "))
        Dikdörtgen_uzun = int(input("Dikdörtgenin uzun kenarını giriniz: "))
        dikdörtgen_alan = Dikdörtgen_kısa * Dikdörtgen_uzun
        print("Dikdörtgenin alanı:", dikdörtgen_alan)
    elif secim == '2':
        print("kare alan")
        kare_kenarlar = int(input("kare kenarını giriniz: "))
        kare_alan = kare_kenarlar * kare_kenarlar
        print("kare alanı:", kare_alan)
    elif secim == '3':
        print("üçgen alan")
        taban_ucgen = int(input("üçgen taban: "))
        yukseklik_ucgen = int(input("üçgen yükseklik: "))
        üçgen_alan = (taban_ucgen * yukseklik_ucgen) / 2
        print("üçgen alanı:", üçgen_alan)
    elif secim == '4':
        print("paralelkenar alan")
        taban_paralelkenar = int(input("paralelkenar taban: "))
        yukseklik_paralelkenar = int(input("paralelkenar yükseklik: "))
        paralelkenar_alan = taban_paralelkenar * yukseklik_paralelkenar
        print("paralelkenar alanı:", paralelkenar_alan)
    elif secim == '5':
        print("silindir alan")
        yarıçap_silindir = int(input("silindir yarıçap: "))
        yukseklik_silindir = int(input("silindir yükseklik: "))
        silindir_alan = 2 * math.pi * yarıçap_silindir * (yarıçap_silindir + yukseklik_silindir)
        print("silindir alanı:", silindir_alan)
    elif secim == '6':
        print("daire alan")
        yarıçap_daire = int(input("daire yarıçap: "))
        daire_alan = math.pi * yarıçap_daire * yarıçap_daire
        print("daire alanı:", daire_alan)
    else:
        print("Geçersiz seçim. Lütfen tekrar deneyin.")



```
