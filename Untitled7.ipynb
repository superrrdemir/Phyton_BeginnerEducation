#Listelerde Farklı İşlemeler
Listeleri temizleme elman seçip aynı anda çıkarma listeyi tersten
yazdırma liste eleman sayısı veya indis bilgisi alma gibi işlemler
yazabiliriz


```python
# Clear komutu -> Her şeyi sil
haftanın_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
haftanın_gunleri.clear()
print(haftanın_gunleri)
```

```python
#count komutu -> Yazılan elemanın sayısı
haftanın_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
print(haftanın_gunleri.count("Pazartesi"))
```

```python
#İndex -> yazılan elemanını elemanın kaçıncı index olduğu
haftanın_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
print(haftanın_gunleri.index("Pazartesi"))
```

```python
#Pop komutu -> listenin içinde bahsedilen indez li elemanları alır yazar ve listeden çıkarırı
haftanın_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
gun = haftanın_gunleri.pop(3)
print(gun)
print(haftanın_gunleri)

```

```python
 # Reverse komutu -> Listeyi TERS ÇEVİRİR
 haftanın_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
 haftanın_gunleri.reverse()
 print(haftanın_gunleri)
```

#Listelerde Arama Yapma ve Döngü Kullanım
Listeler içerisinde belirli koşullar sağlandığında yadırıcak veya işi
yarayacak bilgileri alabilmek için listede arama yapılmalıdır.
Listelerde arama yapmakiçin döngülr kullanılabilir

```python
# 0 - 250 arasında 3'er 3'er sayıları yazdırma
ucerli_liste = []
onbesi_kati =[]
for a in range (0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)
print("Toplam Eleman Sayısı",len(ucerli_liste))
# Bu listenin içinde 5'e bölünen sayıları ekrana yazdırma
for eleman in range(0, len(ucerli_liste)):
  if ucerli_liste[eleman] % 5 == 0:
    onbesi_kati.append(ucerli_liste[eleman])
print(onbesi_kati)
print("Toplam Eleman Sayısı:",len(onbesi_kati))
```
